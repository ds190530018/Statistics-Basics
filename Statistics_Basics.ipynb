Q11. . Explain what a random variable is and differentiate between discrete and continuous random variables.

- A random variable is a numerical value that represents the outcome of a random experiment. It assigns numerical values to events in a sample space.

1. Discrete Random Variable

A discrete random variable takes on a countable number of distinct values.

Examples:

Number of heads in 5 coin flips (0, 1, 2, 3, 4, 5).

Number of customers arriving at a store in an hour.

Number of defective products in a batch.

2. Continuous Random Variable

A continuous random variable takes on an uncountable number of values within an interval. These values come from measuring processes.

Examples:

Height of people (e.g., 170.5 cm, 180.2 cm).

Temperature on a given day (25.4°C, 30.1°C).

Time taken to complete a task (5.3 minutes, 7.8 minutes).
